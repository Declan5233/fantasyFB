```python
all_pos = [QBS, RBS, WRS, TES]

all_pos = [s.strip() for s in all_pos]

for pos in range(len(all_pos)):
    all_pos[pos] = all_pos[pos].split('\n')


for pos in all_pos:
    allPosLoc = []
    dfs = []
    
    # This is for if we want to look at stats from a difrent year
    # for link in range(len(pos)):
    #     pos[link] += '2022/'

    for url in pos:
        print(url)
        page = requests.get(url)
        soup = BeautifulSoup(page.text, 'html')
        try:
            find_team = soup.find('a', class_="nfl-o-cta--link")
            player_team = find_team.text.strip()
        except:
            player_team = ('None')
            print(player_team)
        find_name = soup.find('h1', class_ = "nfl-c-player-header__title")
        player_name = find_name.text.strip()

        find_pos = soup.find('span', class_ = "nfl-c-player-header__position")
        player_pos = find_pos.text.strip()
        
        try:
            table = soup.find_all('table')[1]

            stat_name = table.find_all('th')
            # print(stat_name)

            stat_table_titles = [title.text.strip() for title in stat_name]
            # print(stat_table_titles)


            df = pd.DataFrame(columns = stat_table_titles)

            column_data = table.find_all('tr')
            # print(column_data)

            for row in column_data[1:]:
                row_data = row.find_all('td')
                #print(row_data)
                ind_row_data = [data.text.strip() for data in row_data]

                length = len(df)
                df.loc[length] = ind_row_data

            df['NAME'] = player_name
            

            id_end = url.find('/stats')
            ID = url[28:id_end]
            df['ID'] = ID

            id_start = url.find('logs')
            try:
                year = url[id_start+5:id_start+9]
            except:
                year = datetime.now().year
            df['YEAR'] = year

            df['POSITION'] = player_pos
            
            df['TEAM'] = player_team
            
            
            location = []
            
            for team_shed in df['OPP'].values:
                if '@' in team_shed:
                    location.append('Away')
                else:
                    location.append("Home")
            
            df['LOCATION'] = location
            
            dfs.append(df)
            
        
            
        except:
            continue

    players = pd.concat(dfs, ignore_index = True)
    display(players)
    players.to_csv(f'C:/Users/Decla/pythonStuff/Player_Data_{player_pos}.csv', index = False)
    
```
